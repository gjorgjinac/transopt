```python
import pandas as pd
from sklearn.model_selection import KFold, train_test_split
import numpy as np
import os
```

```python
def split_data(algorithm_name,dim,seed):
    file_name=f'{algorithm_name}_dim_{dim}_seed_{seed}'
    df = pd.read_csv(f'algorithm_run_data/{file_name}.csv',index_col=[0],compression='zip')
    kf = KFold(n_splits=10)
    instance_ids=np.array(df['instance_id'].drop_duplicates())
    output_dir=os.path.join('folds','trajectory_problem_classification',file_name)
    os.makedirs(output_dir, exist_ok=True)
    fold=0
    for train_index, test_index in kf.split(instance_ids):
        train_ids,test_ids=instance_ids[train_index], instance_ids[test_index]
        train_ids,val_ids=train_test_split(train_ids, test_size=0.1)
        print(test_ids)
        for split_ids,split_name in [(train_ids,'train'),(val_ids,'val'),(test_ids,'test')]:
            pd.DataFrame(split_ids).to_csv(os.path.join(output_dir,f'{split_name}_{fold}.csv'))
        fold+=1
    
    
```

```python
for algorithm in ['DE','ES','PSO']:
    for seed in [200,400,600,800,1000]:
        split_data(algorithm,5,seed)
```
