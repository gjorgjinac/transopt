```python
import random

import numpy as np
import pandas as pd
from pymoo.operators.sampling.lhs import LHS
from pymoo.problems import get_problem


def set_random_seed(seed):
    np.random.seed(seed)
    random.seed(seed)
    # tf.random.set_seed(seed)


seed = 1
set_random_seed(seed)


sampling = LHS()
all_populations_df = pd.DataFrame()
sample_size_dimension_factor=100

sample_set_count = 1
for dimension in [3,5,10,20]:
    samples_in_sample_set = sample_size_dimension_factor * dimension
    for problem_id in range(1, 25):
        for instance_id in range(1, 1000):
            problem_name = f'bbob-{problem_id}-{instance_id}'
            problem = get_problem(problem_name, n_var=dimension)
            print(f'{problem_id}_{instance_id}')
            for sample_set_id in range(1, sample_set_count+1):
                sample_set = sampling.do(problem, samples_in_sample_set)

                x_y = []
                for set_id, s in enumerate(sample_set):
                    y = problem.evaluate(s.X)
                    x_y.append(list(s.X) + list(y))
                sample_set_df = pd.DataFrame(x_y)
                sample_set_df['problem_name'] = problem_name
                sample_set_df['problem_id'] = problem_id
                sample_set_df['instance_id'] = instance_id
                sample_set_df['sample_set_id'] = sample_set_id
                all_populations_df =pd.concat([ all_populations_df,sample_set_df])

all_populations_df.to_csv(f'data/bbob_lhs_dim_{dimension}_samples_{sample_size_dimension_factor}d.csv')

```

```python
import pickle

import numpy as np
from cocoex import Suite
from smt.sampling_methods import LHS

suite = Suite('bbob', "instances: 1-999", '')

dataframe = {}
dimension = 3
sample_count_dimension_factor=100
xlimits = np.array([[-5,5] for _ in range(0,dimension)])
sampling = LHS(xlimits=xlimits)


sample_count = sample_count_dimension_factor * dimension
samples = sampling(sample_count)

for f_id in range(1, 25):
    print(f_id)

    for instance in range(1,2):
        f = suite.get_problem_by_function_dimension_instance(f_id, dimension, instance)
        print(f.info)
        for x_index, x in enumerate(samples):
            x_y = list(x).copy()
            x_y.append(f(x))
            dataframe[(f.id_function, f.id_instance, dimension, x_index)] = x_y



pickle.dump( dataframe, open( f"data/lhs_samples_dimension_{dimension}_{sample_count_dimension_factor}_samples.p", "wb" ) )
```
